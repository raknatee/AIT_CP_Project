```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load data
data = np.load('../data/raw_data/p3.npy')
```

```python
#Set parameters

sampling_frequency = 128.0

sampling_interval = 1/sampling_frequency

delta_time = 0.5
```

```python
time_series = np.arange(0,delta_time,sampling_interval)
```

```python
#plotting one row
data_0 = data[0]

figure, ax = plt.subplots(16,1,figsize=(20,20))

for i,data_i in enumerate(data_0):
    ax[i].plot(time_series,data_0[i])
```

```python
data.shape
```

```python
# set the current sample and channel
sample = 0
channel = 0
signal = data[sample,channel,:]

#standardize the data
mean = np.mean(signal)
std = np.std(signal)
signal = (signal-mean)/std


#perform magic
fourier_transform = np.fft.fft(signal)/len(signal)
fourier_transform = np.abs(fourier_transform[:fourier_transform.size//2])

#create a frequency series 
frequency_series = np.arange(fourier_transform.shape[0])/delta_time

print(fourier_transform.shape)
```

```python
figure = plt.figure(figsize=(10,5))
plt.plot(frequency_series,fourier_transform)
```

```python
data_transformed = np.zeros((data.shape[0],data.shape[1],data.shape[2]//2))
```

```python
for sample in range(data.shape[0]):
    for channel in range(data.shape[1]):
        # set the current sample and channel
        signal = data[sample,channel,:]

        #standardize the data
        mean = np.mean(signal)
        std = np.std(signal)
        signal = (signal-mean)/std

        #perform magic
        fourier_transform = np.fft.fft(signal)/len(signal)
        fourier_transform = np.abs(fourier_transform[:fourier_transform.size//2])

        data_transformed[sample,channel,:] = fourier_transform
```

```python
print(data_transformed.shape)
```

```python
data_0 = data_transformed[0]

figure, ax = plt.subplots(16,1,figsize=(20,20))

for i,data_i in enumerate(data_0):
    ax[i].plot(frequency_series,data_0[i])
```
